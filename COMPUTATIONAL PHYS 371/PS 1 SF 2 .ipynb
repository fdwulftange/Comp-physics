```python
from numpy import random
import numpy as np 
import matplotlib.pyplot as plt
```

```python
darts = 100000
circlex = []
circley = []

squarex = []
squarey = [] 

i = 1 
while i<= darts:
    x  = random.uniform(-1,1)
    y  = random.uniform(-1,1)
    if (x**2 + y**2 <= 1):
        circlex.append(x)
        circley.append(y)
    else:
        squarex.append(x)
        squarey.append(y)
    i += 1
    
pi = 4*len(circlex)/float(darts)

print("Pi is approximately:",pi)   
```

```python
plt.plot(circlex,circley,'b.')
plt.plot(squarex,squarey,'m.')
```

```python
def range_darts(low, high): 
    range_darts = np.linspace(low,high)
    diff_funk_N = []
    ratios = []
    for i in range(len(range_darts)):
        num = 1 
        circle = 0
        square = 0
        
        while num<= range_darts[i]:
            x  = random.uniform(-1,1)
            y  = random.uniform(-1,1)
            if (x**2 + y**2 <= 1):
                circle += 1
                square += 1
                
            else:
                square += 1
            num += 1
        
        pi = 4*circle/range_darts[i]
        ratio = square/circle
        value = ratio - 4/np.pi
        ratios.append(ratio)
        diff_funk_N.append(value)
        
    return range_darts,diff_funk_N
```

```python
x,y = range_darts(100,100000)
plt.plot(x,y)
```

```python
def range_darting(low, high): 
    
    range_darts = np.linspace(low,high)
    diff_funk_N = []
    for i in range(len(range_darts)):
        function = 0
        square = 0 
        
        num = 1 
        while num<= range_darts[i]:
            x  = random.uniform(0,1)
            y  = random.uniform(0,1)
            if y <= ((np.exp(-(x)**(1/2)))):
                function += 1
                square += 1
            else:
                square +=1
            num += 1

        integral = function/square
    
        diff_funk_N.append(integral)
        
    return range_darts,diff_funk_N
```

```python
x,y = range_darting(100,100000)
```

```python
def trap(f,x): 
    integral = 0
    for i in range(1,len(f)):
        integral += 0.5 * (f[i] + f[i-1]) * (x[i] - x[i-1])
    return integral

def evaluate(N): 
    f = [] 
    x = np.linspace(0,1,N)
    for i in range(len(x)):
        funk = np.exp(-(x[i])**(1/2))
        f.append(funk)
    return f,x

def plot(low,high): 
    y_vals = [] 
    for i in range(low, high):
        f,x = evaluate(i)
        integral_value = trap(f,x)
        y_vals.append(integral_value)
    ranges  = np.linspace(low, high,len(y_vals))
    return ranges, y_vals
```

```python
x_val,y_val = plot(100,1000)
plt.plot(x_val,y_val)
```
