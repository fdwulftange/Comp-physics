```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 10
N = 1000
x = np.linspace(0,L, N)
squiggle = np.zeros(len(x))

def squiggle_funk(x):
    for i in range(len(x)):
        if x[i] < L/3: 
            squiggle[i] = 3*x[i]/L
        elif x[i] <= 2*L/3: 
            squiggle[i] = 1
        else: 
            squiggle[i] = -3*x[i]/L + 3

    return squiggle

dependent = squiggle_funk(x)

plt.figure(figsize=(10,7))
plt.plot(x, dependent,label = 'Shape of string at t = 0')
plt.legend()
plt.xlabel('String of Boundary - x', size ='20' )
plt.ylabel('String of Shape \u03BE(x, t = 0)', size = '20')
plt.show()
```
